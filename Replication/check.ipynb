```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.zeros(100)

X[0] = np.random.normal(0, 1)

for i in range(1, 100):
    X[i] = np.random.normal(0.9 * X[i-1], 1)
    

Y = 2.5 * X + np.random.normal(0, 1, 100)
```

```python
plt.figure(figsize = (10, 6))

plt.scatter(np.arange(len(Y)), Y)
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
